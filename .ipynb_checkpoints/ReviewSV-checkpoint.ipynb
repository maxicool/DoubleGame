 Consider the curve traced out by the position vector
r(t) = 2t i + 3 sin (2t) j + 3 [ cos (2t) − 1] k.
As the parameter t increases from zero, the curve eventually passes through the point P.
The length along the curve from (0, 0, 0) to P is
π√10/3. What are the coordinates of P

```python
from sympy.abc import x, y, z, p
x = 2 * t
y = 2 * sin(2*t)
z = 3 * (cos(2*t) -1)
I = x.diff(t) ** 2 + y.diff(t) ** 2 + z.diff(t) ** 2
E1 = integrate(sqrt(I),(t,0,p)) - pi * sqrt(10) / 3
s = solve(E1, p)
```
